## Advance operations with business problems

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure: Resturant id, name, location, cuisine type, speciality, rating, number of reviews

data = np.array([
    [3456, 'D Bapi', 'Barrackpore', 'Indian', 'Biryani', 4.5, 120],
    [6754, 'Chowman', 'Ballygunge', 'Chinese', 'Noodles', 4.2, 85],
    [2345, 'Aminia', 'Park Street', 'Indian', 'Kebabs', 4.8, 200],
    [9876, 'The Belgian Waffle Co.', 'Bhowanipure', 'Desserts', 'Waffles', 4.1, 60],
    [4567, 'The French Loaf', 'Salt Lake', 'Bakery', 'Pastries', 4.3, 90],
    [1223, 'Karims', 'Sovabazar', 'Indian', 'Mutton Korma', 4.6, 150],
    [5521, 'Biryani Blues', 'Gariahat', 'Indian', 'Biryani', 4.4, 110],
    [6665, 'The Chinese Wok', 'New Town', 'Chinese', 'Dim Sum', 4.0, 70],
    [2121, 'Asian Spice', 'Salt Lake', 'Asian', 'Stir Fry', 4.2, 95],
    [2321, 'The Great Kabab Factory', 'Park Street', 'Indian', 'Kebabs', 4.7, 180]
])

print("=== Restaurant Data ===")
print("The details of first 4 restaurants are:\n", data[0:4])
```

```python
# Restaurant id, 2022, 2023, 2024, 2025
sales_data = np.array([
    [3456, 50000, 60000, 70000, 80000],
    [6754, 40000, 45000, 50000, 55000],
    [2345, 80000, 90000, 100000, 110000],
    [9876, 30000, 35000, 40000, 45000],
    [4567, 35000, 40000, 45000, 50000],
    [1223, 60000, 70000, 80000, 90000],
    [5521, 55000, 60000, 65000, 70000],
    [6665, 25000, 30000, 35000, 40000],
    [2121, 45000, 50000, 55000, 60000],
    [2321, 20000, 25000, 30000, 35000],
])

print("=== Sales Data ===")
```

```python
# total sales per year
total_sales_per_year = np.sum(sales_data[:, 1:], axis=0)
print("Total sales per year for each restaurant:\n", total_sales_per_year)

# minimum and maximum sales per restaurant
min_sales_per_restaurant = np.min(sales_data[:, 1:], axis=1)
max_sales_per_restaurant = np.max(sales_data[:, 1:], axis=1)
print("Minimum sales per restaurant:\n", min_sales_per_restaurant)
print("Maximum sales per restaurant:\n", max_sales_per_restaurant)

# average sales per restaurant and per year
average_sales_per_restaurant = np.mean(sales_data[:, 1:], axis=1)
average_sales_per_year = np.mean(sales_data[:, 1:], axis=0)
print("Average sales per restaurant:\n", average_sales_per_restaurant)
print("Average sales per year:\n", average_sales_per_year)

# monthly average sales per restaurant
monthly_average_sales_per_restaurant = sales_data[:, 1:] / 12
print("Monthly average sales per restaurant:\n", monthly_average_sales_per_restaurant)
```

```python
# cumulative sales per restaurant
cumulative_sales_per_restaurant = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sales per restaurant:\n", cumulative_sales_per_restaurant)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumulative_sales_per_restaurant, axis=0))
plt.title("Average Cumulative Sales Over Years")
plt.xlabel("Years (2022-2025)")
plt.ylabel("Average Cumulative Sales")
plt.grid(True)
plt.show()
```

## Vector operations

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([5, 4, 3, 2, 1])
print("Vector addition:", vector1 + vector2)
print("Vector subtraction:", vector1 - vector2)
print("Vector multiplication:", vector1 * vector2)
print("Dot product:", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between vectors (in radians):", angle)
```
