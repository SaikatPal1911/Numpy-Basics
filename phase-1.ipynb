## Numpy array and basics

```python
import numpy as np
```

## Creating array from list

```python
array_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", array_1d)
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", array_2d)
```

## List vs numpy array

```python
py_list = [1, 2, 3]
print("Python Multiplication List:", py_list * 2)

np_array = np.array([1, 2, 3])
print("NumPy Multiplication Array:", np_array * 2)
```

## Creating array from scratch

```python
zeros = np.zeros((3, 4))
print("Zeros Array:\n", zeros)

ones = np.ones((2, 3))
print("Ones Array:\n", ones)

full = np.full((2, 2), 7)
print("Full Array:\n", full)

random = np.random.random((2, 3))
print("Random Array:\n", random)

sequence = np.arange(0, 10, 2)
print("Sequence Array:", sequence)
```

## Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector:", vector)

matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\n", matrix)

tensor = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("Tensor:\n", tensor)
```

## Array properties

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Original Array:\n", arr)
print("Shape:", arr.shape)
print("Size:", arr.size)
print("Dimensions:", arr.ndim)
print("Data Type:", arr.dtype)
```

## Array reshaping

```python
arr = np.arange(12)
print("Original Array:", arr)

reshaped_arr = arr.reshape((3, 4))
print("Reshaped Array (3x4):\n", reshaped_arr)

resized_arr = arr.reshape((2, 2, 3))
print("Resized Array (2x2x3):\n", resized_arr)

flattened_arr = reshaped_arr.flatten()
print("Flattened Array:", flattened_arr)

raveled_arr = reshaped_arr.ravel()
print("Raveled Array:", raveled_arr)

transposed_arr = reshaped_arr.T
print("Transposed Array:\n", transposed_arr)
```
